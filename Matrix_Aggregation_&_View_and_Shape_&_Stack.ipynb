# View and Shape

view() and shape() are used to change the shape of the tensors. The core idea  is to transform a tensor into a new shape while keeping the data consistent.

view()  --> Works only if the tensor is stored contiguosly in memory

reshape() --> handle non contiguosly data also.
non contiguos tensor by creating a new tensor if needed.

rehape has more flexibility

view is constrained to use only the contiguos data.

Contigous

original_Data :[[1,2,3],[4,5,6]]

pytorch will store the data in sequential order(from left to right). It look like [1,2,3,4,5,6]  (5*5 grid like structure)

Non Contigous data: Transpose/slicing(Changing the tensor by some operations)

original_Data :[[1,2,3],[4,5,6]]

transposed_data: [[1,4],[2,5],[3,6]]

in memory it will stored as [1,2,3,4,5,6] even we changed the order sill it is stored as the above format.

The data is remains same in the memory, but the access pattern chnages  -- It is the issue. Now pytorch needs to perform indexing which is expensive

Perfomance is high in contiguous data. Non Contiguous data is low performance compared to contiguos data

```python
"""
reshape gets the non contiguous data it wil create new copy which will come to rshpae and that data will be contiguos
if contigous data into reahpe it will just do the reshape

View only with contiguous data.. If u want make our system more strict towards our contiguos data, then we use this operation
"""
```

# Stack Operation

combining the multiple tensor along a new dimension.

Tensor 1= 4 elements and Tensor 2 = 4 elements both tensor have dimension=1

if we need to get all the elements of dim=0 (accessing rows) dim=1 (accesing cols)

1*4 and 1*4 2 tensor total 8 elements

method1: along rows to do this we need to create new dimension (2,4)

tensor1
tensor2

method 2: along cols(4,2)

tensor1 tensor2

```python
import torch
```

```python
tensor1=torch.tensor([1,2,3])
tensor2=torch.tensor([4,5,6])

print(tensor1)
print(tensor2)

print("="* 40)

print(tensor1.ndim)
print(tensor2.ndim)

print("="*40)

print(tensor1.size())
print(tensor2.size())
```

```python
Row_stack=torch.stack([tensor1,tensor2],dim=0)
print(Row_stack)

print(Row_stack.ndim)

print(Row_stack.size())
```

```python
Col_stack=torch.stack([tensor1,tensor2],dim=1)
print(Col_stack)

print(Col_stack.ndim)

print(Col_stack.ndim)

print(Row_stack.size())
```

```python
tensor3=torch.tensor([7,8,9])
tensor4=torch.tensor([10,11,12])

```

```python
Row_stack_new=torch.stack([tensor1,tensor2,tensor3,tensor4],dim=0)
print(Row_stack_new)

print(Row_stack_new.ndim)

print(Row_stack_new.size())
```

```python
Col_stack_new=torch.stack([tensor1,tensor2,tensor3,tensor4],dim=1)
print(Col_stack_new)

print(Col_stack_new.ndim)

print(Col_stack_new.ndim)

print(Row_stack_new.size())
```

```python
# Stacking 2d Tensors

tensor_1=torch.tensor([[1,2,3],[4,5,6]])
tensor_2=torch.tensor([[7,8,9],[10,11,12]])
print(tensor_1.ndim)
print(tensor_2.ndim)
stacked_tensor_2d=torch.stack([tensor_1,tensor_2],dim=0)
print(stacked_tensor_2d)
print(stacked_tensor_2d.ndim)
print(stacked_tensor_2d.shape)
print(stacked_tensor_2d.numel())
```
