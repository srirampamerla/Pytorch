# Tensor Data types

#Data types for tensors

* floating -- for most deep learning tasks ---> float32,float64,float16  -- going with more precision perfomrance/memory will takes larger 16 << 32 << 64
* integers -- for categorical data and indices ---> int32,int64,int8
* booleans -- mask or logical operation
* Complex number -- advanced computation ---> complex64,complex128

memory consumption : float16 << float32 << float64

computation: lower precision will be faster on gpu

numercial precision: float64 is more precise than float32


```python
import torch
```

```python
default_tensor=torch.tensor([1.5,2.5,3.5])
default_tensor.dtype
```

```python
float_tensor=torch.tensor([1.5,2.5,3.5],dtype=torch.float64)
float_tensor.dtype
```

# Impact of Data type on Memory

```python
float16_tensor=torch.ones(1000,dtype=torch.float16)
float32_tensor=torch.ones(1000,dtype=torch.float32)
float64_tensor=torch.ones(1000,dtype=torch.float64)

# tensor.element_size() # It gives size of One element in bytes

#tensor.nelement()  --> Gives the total numbers of elements in tensor
```

```python
print("Memory used by float 16 tensor",float16_tensor.element_size())
print("Memory used by float 32 tensor",float32_tensor.element_size())
print("Memory used by float 64 tensor",float64_tensor.element_size())
```

```python
print("Memory used by float 16 tensor for",float16_tensor.element_size()*float16_tensor.nelement(),"bytes")
print("Memory used by float 32 tensor",float32_tensor.element_size()*float32_tensor.nelement(),"bytes")
print("Memory used by float 64 tensor",float64_tensor.element_size()*float64_tensor.nelement(),"bytes")
```

 float32 --> default for Most NN models

 flaot64 --> high precision computation

 int32  --> General purpose integer

 int64  --> tensor indices

 int8 , int16 --> Reduce memory usage

 bool --> mask and logical operation

float16  --> half precision of your default floating dtype, reduce memory usage

complex64 --> advanced computation
